```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris["data"], columns=iris["feature_names"])
y = pd.DataFrame(iris["target"], columns=["label"])
data = pd.concat([X,y], axis=1)
data.head()
```

## Outline

There are many kinds of graphs in the package, matplotlib.

For example, line plot (plot), scatterplot (scatter), and bar plot (bar).

We take the plot as an example for describing the instructions.

`plt.plot(x, y, fmt, data)`

"fmt" will be replaced by some string like "ro", it means red color + point o

"data" is used in abbreviating x and y to label, then this function will automatically find data in the given data

Other optional parameters:

`color = "b", "g", "r", "c", "m", "y", "k", "w"`

`linestyle = "-", "--", "-.", ":"`

`linewidth = 1`

`marker = "o"`

ref:https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html


## Example: plt.plot

```python
# it will show the text
# the plot means the line plot
plt.plot(data["sepal length (cm)"])
```

```python
# use show() to show the diagram without any words
plt.plot(data["sepal length (cm)"])
plt.show()
```

```python
# 2-dim line plot
# this is meaningless, because the data, iris, is not a time sequence
plt.plot(data["sepal length (cm)"], data["sepal width (cm)"], color="green", marker="o")
```

```python
# this kind of plot is clearer than the prior one 
plt.plot(data["sepal length (cm)"], data["sepal width (cm)"], "go")
```

```python
plt.plot([1,2,6], [5,3,8], color="r", linewidth=2, linestyle="--", label="line1")
plt.plot([1,2,6], [6,4,9], color="b", linewidth=1, linestyle="-.", label="line2")

# Before we use the function legend, we need to give the label in the plot
# the options for parameter, loc, are best, upper right/left, lower right/left, right,
#                                     center right/left, center, lower/upper center
plt.legend(loc="best")
```

```python
plt.plot([1,2,6], [5,3,8], color="r", linewidth=2, linestyle="--", label="line1")
plt.plot([1,2,6], [6,4,9], color="b", linewidth=1, linestyle="-.", label="line2")
plt.legend(loc="best")

# the label for x-axis and y-axis
plt.xlabel("x")
plt.ylabel("y")

# the limitations of the range for x-axis and y-axis
plt.xlim([0, 7])
plt.ylim([2, 10])

# the customized sticks
plt.xticks([1,2,6])
#plt.xticks([1,2,6],["a","b","c"])
#plt.yticks([3,4,5,6,8,9])
plt.yticks([3,4,5,6,8,9],["A1","A2","B1","B2","C1","C2"])

# add a vertical and horizontal line 
for x0 in [1,2,6]:
    plt.axvline(x=x0)
plt.axhline(y=3)
```

## Example: plt.scatter

```python
# ref: https://morvanzhou.github.io/tutorials/data-manipulation/plt/3-1-scatter/

n = 1000
X = np.random.normal(0, 1, n) 
Y = np.random.normal(0, 1, n) 
T = np.arctan2(Y,X)
```

```python
# s is the size, c is the color, alpha is the transparency
plt.scatter(X, Y, s=100, c=T, alpha=.5)
plt.xlim([-1,1])
plt.ylim([-1,1])
```

## Example: plt.bar

```python
# ref: https://morvanzhou.github.io/tutorials/data-manipulation/plt/3-2-Bar/

n = 6
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
```

```python
plt.bar(X, +Y1)
plt.bar(X, -Y2)

plt.xlim(-0.5, n-0.5)
plt.xticks(())

plt.ylim(-1.25, 1.25)
plt.yticks(())
```

```python
# the color can refer to this website, https://www.toodoo.com/db/color.html
plt.bar(X, +Y1, facecolor="#880000", edgecolor="white")
plt.bar(X, -Y2, facecolor="#003377", edgecolor="white")

plt.xlim(-0.5, n-0.5)
plt.xticks(())

plt.ylim(-1.25, 1.25)
plt.yticks(())

# ha means horizontalalignment; va means the verticalalignment 
for x, y in zip(X, Y1):
    # ha: horizontal alignment
    # va: vertical alignment
    plt.text(x, y + 0.05, "%.2f" % y, ha="center", va="bottom")

for x, y in zip(X, Y2):
    # ha: horizontal alignment
    # va: vertical alignment
    plt.text(x, -y - 0.05, "%.2f" % y, ha="center", va="top")
```
