```python
iris = datasets.load_iris()
X = pd.DataFrame(iris["data"], columns=iris["feature_names"])
y = pd.DataFrame(iris["target"], columns=["label"])
data = pd.concat([X,y], axis=1)
data.head()
```

## Example: sns.distplot

```python
sns.set()
sns.distplot(data["sepal length (cm)"])
```

```python
# without kernel density estimation line
sns.set()
sns.distplot(data["sepal length (cm)"], kde=False)
```

```python
# transpose the plot
sns.set()
sns.distplot(data["sepal length (cm)"], vertical=True)
```

## Example: sns.jointplot and sns.pairplot

```python
sns.set()
sns.jointplot(data["sepal length (cm)"], data["sepal width (cm)"])

# another format
# sns.jointplot(x = "sepal length (cm)", y = "sepal width (cm)", data = data)
```

```python
sns.set()
sns.pairplot(data[["sepal length (cm)", "sepal width (cm)"]])
```

```python
# use parameter, alpha, in the plot_kws to adjust the transparency
sns.set()
sns.pairplot(data[["sepal length (cm)", "sepal width (cm)"]], plot_kws={"alpha": 0.5})
```

```python
# use the parameter, kind, to fit a regression line
# and the plot_kws to change the color of the line and the transparency of the points
sns.set()
sns.pairplot(data[["sepal length (cm)", "sepal width (cm)"]], kind="reg",
             plot_kws={"line_kws":{"color":"red"}, "scatter_kws":{"alpha": 0.5}})
```

## Example: sns.factorplot

```python
np.random.seed(505)
sample = np.random.randint(0, 5, 10)
df = pd.DataFrame({"x":range(10), "y":sample})

sns.set()
sns.factorplot(x = "x", y = "y", data=df)
```

```python
# the paramter, kind, can change the type of points
# there are the options: {point, bar, count, box, violin, strip}

np.random.seed(505)
sample = np.random.randint(0, 5, 10)
df = pd.DataFrame({"x":range(10), "y":sample})

sns.set()
sns.factorplot(x = "x", y = "y", data=df, kind="bar")
```

## Example: sns.countplot

```python
sns.set()
sns.countplot(x = "label", data = data)
```

## Example: sns.boxplot

```python
sns.set()
#sns.boxplot(data["sepal length (cm)"])
sns.boxplot(x="sepal length (cm)", data=data)
```

```python
sns.set()
sns.boxplot(x="label", y="sepal length (cm)", data=data)
```
